```python
#1
def smart_parking(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        # Track peak usage
        peak = max(peak, parked)

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full (Over Capacity Alert!)")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)


# Sample Input
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
#2
def estimate_delivery_time(distance, traffic, weather):
    # Base time (5 min per km)
    time = distance * 5

    # Traffic impact
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    else:
        time += 5

    # Weather impact
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```
